```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Poisson distribution

```python
from scipy.stats import poisson
```

```python
A = [1, 5, 10, 100]
distributions = []
for a in A:
    distribution = poisson.rvs(a, size=10000)
    distributions.append(distribution)

for distribution in distributions:
    plt.hist(distribution, bins=150)
    plt.show()
```

```python
mu = 10
x = np.arange( poisson.ppf(0.01, mu), poisson.ppf(0.99, mu)) 
plt.figure(figsize=(12, 6))
plt.plot(x, poisson.pmf(x, mu))
plt.title('Poisson PMF', fontsize=20)

# pmf Probability mass function

plt.vlines(x, 0, poisson.pmf(x, mu), lw=5) #vertical lines
```

## Normal distribution

### Probability density

```python
mu, sigma = 0.5, 0.1

lin = np.linspace(0, 1, 100)
plt.plot(lin, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp(-(lin - mu)**2 / (2 * sigma**2) ), linewidth=3, color='y')

plt.show()
```

```python
from scipy.stats import norm
```

```python
x = norm.rvs(mu, sigma, size=1000)
```

```python
x.mean()
```

```python
x_squares = x**2
```

```python
x_squares.mean()
```

```python
norm.pdf(0.5, loc=mu, scale=sigma)
```

```python
norm.pdf(0.5, loc=mu, scale=sigma)
```

```python
plt.plot(np.linspace(0, 1, 100), norm.pdf(np.linspace(0, 1, 100), mu, sigma))
plt.plot(np.linspace(0, 1, 100), norm.pdf(np.linspace(0, 1, 100), mu, sigma) ** 2)
```

#### Variance 

```python
x_squares.mean() - x.mean()**2 
```

```python
x.var()
```

#### Normalization

```python
 x_1 = (x - x.mean())/ x.std()
```

```python
x_1.std()
```

```python
x_1.mean()
```

#### Standatd Deviation (std)

```python
np.sqrt(x.var())
```

```python
x.std()
```

#### Asymmetry 

```python
from scipy.stats import moment

moment(x, moment=3) / x.std()**3
```

```python
from scipy.stats import skew
```

```python
skew(x)
```

#### Kurtosis

```python
from scipy.stats import kurtosis
```

```python
moment(x, moment=4) / x.std()**4 - 3
```

```python
kurtosis(x) 
```

## Chi-square distribution 

```python
x1 = norm.rvs(size=1000)
x2 = norm.rvs(size=1000)
x3 = norm.rvs(size=1000)
```

```python
plt.hist(x1, alpha=0.1, color='y')
plt.hist(x2, alpha=0.1, color='b')
plt.hist(x3, alpha=0.1, color='r')
plt.show()
```

```python
x = x1**2 + x2**2 + x3**2
```

```python
plt.hist(x, bins=100)
plt.show()
```

```python
from scipy.stats import chi2
```

```python
x = chi2.rvs(df=6, size=1000)
```

```python
plt.hist(x, bins=100)
plt.show()
```

```python
df = 2
fig, ax = plt.subplots(1, 1)
x = np.linspace(chi2.ppf(0.01, df),
                chi2.ppf(0.99, df), 100)
ax.plot(x, chi2.pdf(x, df),
       'r-', lw=5, alpha=0.6, label='chi2 pdf')
plt.show()
```

## Student's t-distribution

```python
from scipy.stats import t
```

```python
x = t.rvs(df=2, size=1000)
```

```python
plt.hist(x, bins=100)
plt.show()
```

```python
df = 2
fig, ax = plt.subplots(1, 1)

x = np.linspace(t.ppf(0.01, df),
                t.ppf(0.99, df), 100)
ax.plot(x, t.pdf(x, df),
       'r-', lw=5, alpha=0.6, label='t pdf')
plt.show()
```
